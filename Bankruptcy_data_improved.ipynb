### Decision Tree without PCA

```python
from sklearn import tree
from sklearn.model_selection import StratifiedKFold
```

```python
x_train, x_test, y_train, y_test = train_test_split(scaled_data, y, test_size = 0.3, random_state= 42)
```

```python
i = 1
kf = StratifiedKFold(n_splits = 5, random_state = 1, shuffle = True)
mean_score = 0
for train_index, test_index in kf.split(scaled_data,y):
    print('\n{} of Kfold {}'.format(i,kf.n_splits))
    xtr,xvl = x.iloc[train_index],x.loc[test_index]
    ytr,yvl = y.iloc[train_index], y.loc[test_index]
    model = tree.DecisionTreeClassifier(random_state = 1)
    model.fit(xtr,ytr)
    pred_test = model.predict(xvl)
    score = precision_score(yvl,pred_test)
    print("precision Score", score)
    i+=1
    pred_test = model.predict(x_test)
    mean_score += score/5
print ('\nMean precision score is {}' .format(mean_score))
```

```python
print(classification_report(y_test,pred_test))
```

```python
confusion_matrix(y_test,pred_test)
```

Since precision and recall is very poor for class 1 we will do hyperparameter tuning using GridSearchCV

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from random import randint
```

```python
param_grid = {'max_depth': range(3, 10, 2),
              'min_samples_leaf': range(50, 150, 20),
              'min_samples_split': range(50, 150, 20),
              'criterion': ["entropy", "gini"],
              'max_features':["auto", "sqrt", "log2"],
              'class_weight':['balanced']
             }

n_folds = 3
              
# Instantiate the grid search model
dtree = DecisionTreeClassifier()
grid_search = GridSearchCV(estimator=dtree, param_grid=param_grid, cv=n_folds, verbose=1, n_jobs=-1)

# Fit the grid search to the data
grid_search.fit(x_train,y_train)
```

```python
pred_test = grid_search.predict(x_test)
```

```python
confusion_matrix(y_test,pred_test)
```

```python
print(classification_report(y_test,pred_test))
```

```python
# printing the optimal accuracy score and hyperparameters
print("best accuracy", grid_search.best_score_)
print(grid_search.best_estimator_)
```
